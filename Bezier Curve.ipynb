```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define control points (x,y)
p0 = np.array([0, 0])       # start point
p1 = np.array([500, 500])   # control point
p2 = np.array([900, 1200])  # end point

# Generate curve points
curve_points = []
for t in np.linspace(0, 1, 200):  # 200 points along the curve
    point = (1-t)**2 * p0 + 2*(1-t)*t * p1 + t**2 * p2
    curve_points.append(point.astype(int))

curve_points = np.array(curve_points)

# Draw curve on image
for i in range(len(curve_points)-1):
    cv2.line(rgb, tuple(curve_points[i]), tuple(curve_points[i+1]), (255,0,0), 2)

# Show result
plt.imshow(rgb)
plt.title("Quadratic Bézier Curve")
plt.axis("off")
plt.show()
```
