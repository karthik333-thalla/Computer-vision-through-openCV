```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img1 = cv2.imread(r"C:\Users\user\OneDrive\Desktop\faces.webp")

# Convert to grayscale
gray_img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Show grayscale image
cv2.imshow("Original_grayscale_image", gray_img)
cv2.waitKey(0)

# Load Haar Cascade
haar_face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_alt.xml"
)

# Detect faces
faces = haar_face_cascade.detectMultiScale(
    gray_img,
    scaleFactor=1.1,
    minNeighbors=5
)

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img1, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the detected image
cv2.imshow("Final_detected_image", img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
