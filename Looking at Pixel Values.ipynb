```python
import cv2
import pandas as pd

# Load the image
img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")

if img is None:
    print("Image not found!")
else:
    # Flatten the image into a 1D array
    pixel_values = img.flatten()

    # Convert pixel values to a DataFrame
    df = pd.DataFrame(pixel_values)

    # Save as CSV instead of Excel
    filepath = 'pixel_values1.csv'
    df.to_csv(filepath, index=False)

    print(f"Pixel values saved to {filepath}")
```
