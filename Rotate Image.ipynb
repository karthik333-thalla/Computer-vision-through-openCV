```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get image dimensions
(h, w) = rgb.shape[:2]
center = (w // 2, h // 2)

# Rotate image by 20 degrees
M = cv2.getRotationMatrix2D(center, 20, 1.0)
rotated = cv2.warpAffine(rgb, M, (w, h))

# Show rotated image
plt.imshow(rotated)
plt.title("Rotated by 20°")
plt.axis("off")
plt.show()
```
