```python
import cv2
import matplotlib.pyplot as plt

# Load your image
img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")

if img is None:
    print("Image not found!")
else:
    # Convert BGR → RGB for display
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Draw a circle at (500,500) with radius 100
    # Color = red (255,0,0), thickness = -1 (filled circle)
    cv2.circle(rgb, (500,500), 100, (255,0,0), -1)

    # Show the image
    plt.imshow(rgb)
    plt.title("Image with Circle")
    plt.axis("off")
    plt.show()
```
