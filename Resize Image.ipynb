```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize to 100x100
resized = cv2.resize(rgb, (100, 100))

# Save resized image
cv2.imwrite("f1_resized.png", cv2.cvtColor(resized, cv2.COLOR_RGB2BGR))

# Show resized image
plt.imshow(resized)
plt.title("Resized to 100x100")
plt.axis("off")
plt.show()
```
