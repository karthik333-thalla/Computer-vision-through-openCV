```python
import cv2
import matplotlib.pyplot as plt

# Read image
src = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")

if src is None:
    print("Image not found!")
else:
    # Convert to grayscale
    gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

    # Apply histogram equalization
    src_eqlzd = cv2.equalizeHist(gray)

    # Show results
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1); plt.imshow(gray, cmap='gray'); plt.title("Original Grayscale"); plt.axis("off")
    plt.subplot(1,2,2); plt.imshow(src_eqlzd, cmap='gray'); plt.title("Histogram Equalized"); plt.axis("off")
    plt.show()

    # Save equalized image
    cv2.imwrite("f1_equalized.png", src_eqlzd)
```
