```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")

if image is None:
    print("Image not found!")
else:
    # Convert BGR → RGB for display
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Define font
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Add text
    cv2.putText(rgb,
                "WELCOME TO FORMULA 1",   # text
                (50, 100),                # position (x,y)
                font,                     # font type
                2,                        # font scale
                (255, 0, 0),              # color (blue)
                4,                        # thickness
                cv2.LINE_AA)              # line type

    # Show image
    plt.imshow(rgb)
    plt.title("Image with Text")
    plt.axis("off")
    plt.show()
```
