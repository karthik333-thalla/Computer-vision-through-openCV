```python
import cv2

# Read the palm image
image = cv2.imread(r"C:\Users\user\Downloads\WhatsApp Image 2026-06-25 at 10.15.09 AM.jpeg")

# Check image loaded or not
if image is None:
    print("Image not found")
    exit()

# Display original image
cv2.imshow("Original Palm Image", image)
cv2.waitKey(0)


# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow("Gray Palm Image", gray)
cv2.waitKey(0)


# Apply Canny Edge Detection to find palm lines
# Adjust thresholds if required
edges = cv2.Canny(gray, 40, 55, apertureSize=3)

cv2.imshow("Detected Palm Lines", edges)
cv2.waitKey(0)


# Invert colors so detected lines become black
edges = cv2.bitwise_not(edges)

cv2.imshow("Inverted Palm Lines", edges)
cv2.waitKey(0)


# Save detected lines image
cv2.imwrite("palmlines.jpg", edges)


# Read saved image
palmlines = cv2.imread("palmlines.jpg")


# Blend original image with detected lines
final_image = cv2.addWeighted(
    palmlines, 0.3,
    image, 0.7,
    0
)


# Display final output
cv2.imshow("Final Palm Line Detection", final_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
