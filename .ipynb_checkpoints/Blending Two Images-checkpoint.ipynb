```python
import cv2
import matplotlib.pyplot as plt

# Read image 1
img1 = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
# Read image 2
img2 = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")

if img1 is None or img2 is None:
    print("One of the images was not found!")
else:
    # Resize second image to match the first (important for blending)
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

    # Define alpha and beta
    alpha = 0.30  # weight for img1
    beta = 0.70   # weight for img2

    # Blend images
    blended = cv2.addWeighted(img1, alpha, img2, beta, 0.0)

    # Convert BGR → RGB for display
    blended_rgb = cv2.cvtColor(blended, cv2.COLOR_BGR2RGB)

    # Show blended image
    plt.imshow(blended_rgb)
    plt.title("Blended Image")
    plt.axis("off")
    plt.show()
```
