```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")

if image is None:
    print("Image not found!")
else:
    # Convert BGR → RGB for display
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Define dilation sizes
    d1, d2, d3 = 5, 10, 20

    # Define dilation types
    t1 = cv2.MORPH_RECT
    t2 = cv2.MORPH_CROSS
    t3 = cv2.MORPH_ELLIPSE

    # Create structuring elements (kernels)
    kernel1 = cv2.getStructuringElement(t1, (2*d1+1, 2*d1+1), (d1, d1))
    kernel2 = cv2.getStructuringElement(t2, (2*d2+1, 2*d2+1), (d2, d2))
    kernel3 = cv2.getStructuringElement(t3, (2*d3+1, 2*d3+1), (d3, d3))

    # Apply dilation
    dilated1 = cv2.dilate(rgb, kernel1)
    dilated2 = cv2.dilate(rgb, kernel2)
    dilated3 = cv2.dilate(rgb, kernel3)

    # Show results
    plt.figure(figsize=(15,5))
    plt.subplot(1,4,1); plt.imshow(rgb); plt.title("Original"); plt.axis("off")
    plt.subplot(1,4,2); plt.imshow(dilated1); plt.title("Dilation RECT"); plt.axis("off")
    plt.subplot(1,4,3); plt.imshow(dilated2); plt.title("Dilation CROSS"); plt.axis("off")
    plt.subplot(1,4,4); plt.imshow(dilated3); plt.title("Dilation ELLIPSE"); plt.axis("off")
    plt.show()
```
