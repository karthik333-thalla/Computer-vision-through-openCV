```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

yuv = cv2.cvtColor(rgb, cv2.COLOR_RGB2YUV)
yuv_rgb = cv2.cvtColor(yuv, cv2.COLOR_YUV2RGB)

plt.subplot(1,2,1); plt.imshow(yuv); plt.title("YUV"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(yuv_rgb); plt.title("Back to RGB"); plt.axis("off")
plt.show()
```
