```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")

if image is None:
    print("Image not found!")
else:
    # Convert BGR → RGB for display
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Create a dummy image with same shape
    new_image = np.zeros(rgb.shape, rgb.dtype)

    # Brightness and contrast parameters
    contrast = 3.0   # higher = more contrast
    bright = 2       # higher = brighter

    # Apply contrast and brightness adjustment
    for y in range(rgb.shape[0]):
        for x in range(rgb.shape[1]):
            for c in range(rgb.shape[2]):
                new_image[y, x, c] = np.clip(contrast * rgb[y, x, c] + bright, 0, 255)

    # Show original and modified images
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1); plt.imshow(rgb); plt.title("Original"); plt.axis("off")
    plt.subplot(1,2,2); plt.imshow(new_image); plt.title("Contrast & Brightness Adjusted"); plt.axis("off")
    plt.show()
```
