```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")

if image is None:
    print("Image not found!")
else:
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply different thresholding techniques
    _, thresh_binary       = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY)
    _, thresh_binary_inv   = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
    _, thresh_trunc        = cv2.threshold(gray, 50, 255, cv2.THRESH_TRUNC)
    _, thresh_tozero       = cv2.threshold(gray, 50, 255, cv2.THRESH_TOZERO)
    _, thresh_tozero_inv   = cv2.threshold(gray, 50, 255, cv2.THRESH_TOZERO_INV)

    # Display results
    titles = ["Original Gray", "Binary", "Binary Inverted", "Truncated", "To Zero", "To Zero Inverted"]
    images = [gray, thresh_binary, thresh_binary_inv, thresh_trunc, thresh_tozero, thresh_tozero_inv]

    plt.figure(figsize=(15,8))
    for i in range(6):
        plt.subplot(2,3,i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis("off")
    plt.show()
```
