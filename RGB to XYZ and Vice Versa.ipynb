```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

xyz = cv2.cvtColor(rgb, cv2.COLOR_RGB2XYZ)
xyz_rgb = cv2.cvtColor(xyz, cv2.COLOR_XYZ2RGB)

plt.subplot(1,2,1); plt.imshow(xyz); plt.title("XYZ"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(xyz_rgb); plt.title("Back to RGB"); plt.axis("off")
plt.show()
```
