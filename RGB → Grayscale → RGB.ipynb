```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

plt.subplot(1,2,1); plt.imshow(gray, cmap="gray"); plt.title("Grayscale"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(gray_rgb); plt.title("Back to RGB"); plt.axis("off")
plt.show()
```
