```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")

if image is None:
    print("Image not found!")
else:
    # Convert BGR → RGB for display
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply Median Blur (kernel size must be odd)
    median_blur = cv2.medianBlur(rgb, 9)

    # Apply Gaussian Blur (kernel size and sigma)
    gaussian_blur = cv2.GaussianBlur(rgb, (9, 9), 10)

    # Apply Bilateral Filter (diameter, sigmaColor, sigmaSpace)
    bilateral_blur = cv2.bilateralFilter(rgb, 9, 100, 75)

    # Show results
    plt.figure(figsize=(15,5))
    plt.subplot(1,4,1); plt.imshow(rgb); plt.title("Original"); plt.axis("off")
    plt.subplot(1,4,2); plt.imshow(median_blur); plt.title("Median Blur"); plt.axis("off")
    plt.subplot(1,4,3); plt.imshow(gaussian_blur); plt.title("Gaussian Blur"); plt.axis("off")
    plt.subplot(1,4,4); plt.imshow(bilateral_blur); plt.title("Bilateral Blur"); plt.axis("off")
    plt.show()
```
