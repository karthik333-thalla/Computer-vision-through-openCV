```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"C:\Users\user\OneDrive\Desktop\f1 car image -2.webp")
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define kernels
kernel_rect   = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
kernel_cross  = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
kernel_ellipse= cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))

# Apply erosion
eroded_rect    = cv2.erode(rgb, kernel_rect)
eroded_cross   = cv2.erode(rgb, kernel_cross)
eroded_ellipse = cv2.erode(rgb, kernel_ellipse)

# Show results
plt.figure(figsize=(12,4))
plt.subplot(1,4,1); plt.imshow(rgb); plt.title("Original"); plt.axis("off")
plt.subplot(1,4,2); plt.imshow(eroded_rect); plt.title("Erosion RECT"); plt.axis("off")
plt.subplot(1,4,3); plt.imshow(eroded_cross); plt.title("Erosion CROSS"); plt.axis("off")
plt.subplot(1,4,4); plt.imshow(eroded_ellipse); plt.title("Erosion ELLIPSE"); plt.axis("off")
plt.show()
```
