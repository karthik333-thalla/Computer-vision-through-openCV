```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\user\OneDrive\Desktop\F1 car image.png")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

lab = cv2.cvtColor(rgb, cv2.COLOR_RGB2LAB)
lab_rgb = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)

plt.subplot(1,2,1); plt.imshow(lab); plt.title("LAB"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(lab_rgb); plt.title("Back to RGB"); plt.axis("off")
plt.show()
```
